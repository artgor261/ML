```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("plane.jpg", cv2.IMREAD_COLOR)
cv2.imshow("image", img)
cv2.waitKey(100)
cv2.destroyAllWindows()
```

```python
# вывод изображения
img = cv2.imread('plane.jpg', cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

```python
# сохранение изображения
img = cv2.imread('plane.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imwrite('D:\Artem\image.jpg', img)
```

```python
# изменение размера изображения
img = cv2.imread('plane.jpg', cv2.IMREAD_GRAYSCALE)
img_50x50 = cv2.resize(img, (50, 50))
plt.imshow(img_50x50, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# обрезка изображения
img = cv2.imread('plane.jpg', cv2.IMREAD_GRAYSCALE)
cropped_img = img[:,:128]
plt.imshow(cropped_img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# свёртка изображения
img = cv2.imread('plane.jpg', cv2.IMREAD_GRAYSCALE)
img_blurry = cv2.blur(img, (50, 50)) # второй аргумент это ядро свёртки
plt.imshow(img_blurry, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# применение функции filter2D для свёртки
kernel = np.ones((5, 5)) / 25.0
img = cv2.imread('plane.jpg', cv2.IMREAD_GRAYSCALE)
img_filter = cv2.filter2D(img, -1, kernel)
plt.imshow(img_filter, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# увеличение контрастности с помощью выравнивания гистограммы
img = cv2.imread("plane.jpg", cv2.IMREAD_GRAYSCALE)
contrast = cv2.equalizeHist(img)
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# выделение цвета(фильтрация)
img_bgr = cv2.imread("plane.jpg")
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
lower_blue = np.array([50, 100, 50]) # нижняя граница фильтрации
upper_blue = np.array([130, 255, 255]) # верхняя граница фильтрации
filtered_img = cv2.inRange(img_hsv, lower_blue, upper_blue) # фильтрация(выделение только тех пикселей, которые соотвествуют указанному диапазону)
img_rgb = cv2.cvtColor(filtered_img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
# цвета на изображении, удовлетворяющие диапазону, выделяются черным. Пиксели вне диапазона принимают значения 0, 
# пиксели внутри диапазана принимают значения 1
```

```python
# Порогование или шумоподавление изображения
img_grey = cv2.imread("plane.jpg", cv2.IMREAD_GRAYSCALE)
max = 255
neighbours_size = 99
constant = 10
img_binarized = cv2.adaptiveThreshold(img_grey, max, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, neighbours_size, constant)
img_rgb = cv2.cvtColor(img_binarized, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
```

```python
# детектор краев Кэнни
img = cv2.imread('plane.jpg', cv2.IMREAD_GRAYSCALE)
canny_edge = cv2.Canny(img, 100, 200)
plt.imshow(canny_edge)
plt.axis("off")
```
